# create_train_val_test_list

```python
import shutil
import random
import os

path = "./train_data/"

#原始路徑
image_original_path = path+"resize_data/"
label_original_path = path+"txt/"

cur_path = os.getcwd()
```

```python
#訓練集路徑
train_image_path = os.path.join(cur_path,path+"images/train/")
train_label_path = os.path.join(cur_path,path+"labels/train/")

#驗證集路徑
val_image_path   = os.path.join(cur_path,path+"images/val/")
val_label_path   = os.path.join(cur_path,path+"labels/val/")

#測試集路徑
test_image_path  = os.path.join(cur_path,path+"images/test/")
test_label_path  = os.path.join(cur_path,path+"labels/test/")


#訓練集目錄
list_train = os.path.join(cur_path,path+"train.txt")
list_val   = os.path.join(cur_path,path+"val.txt")
list_test  = os.path.join(cur_path,path+"test.txt")


train_percent = 0.8
val_percent   = 0.1
test_percent  = 0.1
```

```python
def del_file(path):
    for i in os.listdir(path):
        file_data = path + "\\" + i
        os.remove(file_data)

def mkdir():
    if not os.path.exists(train_image_path):
        os.makedirs(train_image_path)
    else:
        del_file(train_image_path)
    
    if not os.path.exists(train_label_path):
        os.makedirs(train_label_path)
    else:
        del_file(train_label_path)
    
    if not os.path.exists(val_image_path):
        os.makedirs(val_image_path)
    else:
        del_file(val_image_path)
        
    if not os.path.exists(val_label_path):
        os.makedirs(val_label_path)
    else:
        del_file(val_label_path)
    
    if not os.path.exists(test_image_path):
        os.makedirs(test_image_path)
    else:
        del_file(test_image_path)
        
    if not os.path.exists(test_label_path):
        os.makedirs(test_label_path)
    else:
        del_file(test_label_path)
        
def clearfile():
    if os.path.exists(list_train):
        os.remove(list_train)
    
    if os.path.exists(list_val):
        os.remove(list_val)
    
    if os.path.exists(list_test):
        os.remove(list_test)
```

```python
def main():
    mkdir()
    clearfile()
    
    file_train = open(list_train,'w')
    file_val   = open(list_val,'w')
    file_test  = open(list_test,'w')
    
    total_txt = os.listdir(label_original_path)
    #print(total_txt[0])
    num_txt   = len(total_txt)
    list_all_txt = range(num_txt)
    
    num_train = int(num_txt * train_percent)
    num_val   = int(num_txt * val_percent)
    num_test  = num_txt - num_train - num_val
    #num_test  = 0
    
    
    train = random.sample(list_all_txt,num_train)
    
    val_test = [i for i in list_all_txt if not i in train]
    
    val = random.sample(val_test,num_val)
    
    print("訓練集數目：{},驗證集數目：{},測試集數目：{}".format(len(train),len(val),num_test))
    
    for i in list_all_txt:
        name = total_txt[i][:-4]
        #print(name)
        srcImage = image_original_path + name + '.jpg'
        srcLabel = label_original_path + name + '.txt'
        
        if i in train:
            dst_train_Image = train_image_path + name +'.jpg'
            dst_train_Label = train_label_path + name +'.txt'
            #print(dst_train_Image)
            #print(dst_train_Label)
            shutil.copyfile(srcImage,dst_train_Image)
            shutil.copyfile(srcLabel,dst_train_Label)
            file_train.write(dst_train_Image + '\n')
        
        elif i in val:
            dst_val_Image = val_image_path + name +'.jpg'
            dst_val_Label = val_label_path + name +'.txt'
            shutil.copyfile(srcImage,dst_val_Image)
            shutil.copyfile(srcLabel,dst_val_Label)
            file_val.write(dst_val_Image + '\n')
        
        else:
            dst_test_Image = test_image_path + name +'.jpg'
            dst_test_Label = test_label_path + name +'.txt'
            shutil.copyfile(srcImage,dst_test_Image)
            shutil.copyfile(srcLabel,dst_test_Label)
            file_test.write(dst_test_Image + '\n')
    
    file_train.close()
    file_test.close()
    file_val.close()
```

```python
main()
```

```python
import cv2
import os
import glob
import numpy as np
import pandas as pd
import json

def txt_to_image(txt_path,image_path, save_image_path):
    if not os.path.isdir(save_image_path):
        os.mkdir(save_image_path)
    
    txt_path = txt_path+'*.txt'  #txt\\ 要改成自己的檔案路徑
    txt = glob.glob(txt_path)
    image_path = image_path+'*.jpg'
    image = glob.glob(image_path)
    width = 640
    height = 360

    #save_path = "C:/Users/User/Downloads/image/labels/"
    save_path = save_image_path
    
    for j in range(len(image)):
        img = image[j].split("\\")
        image_name = img[1][:-4]
        #print(img)
        k = 0
        for i in range(len(txt)):
            t = txt[i].split("\\")
            save = t[1][:-4]
            # 製作黑色底圖
            img = np.zeros((360,640,3), np.uint8)
            if image_name == save :
                data = pd.read_table(txt[i],header=None,encoding='utf-8',engine ="python"
                                     ,sep='::')
                #print(img)
                for j in range(len(data)):
                    data_np = data.to_numpy()
                    temp = data_np[j].tolist()
                    t = temp[0].split(" ")
                    #print(temp[0])
                    if t[0] == '0':
                        color = (0,0,255)
                        #print(1)
                    elif t[0] == '1':
                        color = (0,255,0)
                    else:
                        color = (255,0,0)
                        #print(2)
                    #print(float(t[1]))
                    cx = int(float(t[1])*640)
                    cy = int(float(t[2])*360)
                    w  = int(float(t[3])*640)
                    h  = int(float(t[4])*360)
                    left_up = ((cx-w//2),(cy-h//2))
                    right_down = ((cx+w//2),(cy+h//2))
                    #print(left_up,right_down)
                    img = cv2.rectangle(img, left_up, right_down, color, -1)

                # 寫入圖檔
                cv2.imwrite(save_path+save+".jpg", img)
            else:
                k+=1
        if k ==len(txt):
            cv2.imwrite(save_path+image_name+".jpg",img)
            
```

# Calculate_Score

```python
import os
import cv2
import numpy as np
import glob
import pandas as pd
from tensorflow.keras import backend as K

def dice_coef(y_true, y_pred, smooth=0):
    intersection = K.sum(y_true * y_pred, axis=[0,1])
    union = K.sum(y_true, axis=[0,1]) + K.sum(y_pred, axis=[0,1])
    return K.mean((2. * intersection + smooth) / (union + smooth), axis=0)

def Calculate_Score(answer_path, prediction_path):
    answer_path = answer_path
    prediction_path = prediction_path
    answer_list = os.listdir(answer_path)
    prediction_list = os.listdir(prediction_path)
    #print(answer_list)
    #print(prediction_list)
    
    B_score = []
    G_score = []
    R_score = []
    for i in range(len(answer_list)):
        ans = cv2.imread(f'{answer_path}{answer_list[i]}').astype('float32')
        ans[:, :, 0] = np.where(ans[:, :, 0] > 1, 1, 0)
        ans[:, :, 1] = np.where(ans[:, :, 1] > 1, 1, 0)
        ans[:, :, 2] = np.where(ans[:, :, 2] > 1, 1, 0)
        pre = cv2.imread(f'{prediction_path}{prediction_list[i]}').astype('float32')
        pre[:, :, 0] = np.where(pre[:, :, 0] > 1, 1, 0)
        pre[:, :, 1] = np.where(pre[:, :, 1] > 1, 1, 0)
        pre[:, :, 2] = np.where(pre[:, :, 2] > 1, 1, 0)

        
        for j in range(0,3):
            # BGR
            dice = dice_coef(ans[:,:,0+j:1+j], pre[:,:,0+j:1+j])
            dice_score = dice.numpy()
            if np.isnan(dice_score):
                print(prediction_list[i])
            
            if j == 0:
                B_score.append(dice_score)
            elif j == 1:
                G_score.append(dice_score)
            else:
                R_score.append(dice_score)
    print(np.mean(B_score))
    print(np.mean(G_score))
    print(np.mean(R_score))
    print('The final score : ', (np.mean(B_score)+np.mean(G_score)+np.mean(R_score))/3)
```
